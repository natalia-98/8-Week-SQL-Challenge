<p style='font-family:times new roman;font-size:16px'>Update the exclusions and extra columns. Set all rows with empty strings or 'null' values in the exclusions and extra columns to NULL values.</p>

```python
c.execute('''
UPDATE customer_orders_clean
SET exclusions =
    (CASE 
        WHEN exclusions = '' OR exclusions LIKE '%null%'
            THEN NULL 
        ELSE exclusions END)
    , extras = 
    (CASE 
        WHEN extras = '' OR extras LIKE '%null%' OR extras is NULL
            THEN NULL
        ELSE extras END)

''')
conn.commit()
```

```python
customer_orders_clean = pd.read_sql('''
SELECT * FROM customer_orders_clean
;''',conn)
print('Table: customer_orders_clean')
customer_orders_clean
```

<p style='font-family:times new roman;font-size:16px;'>Create a copy of the runner_orders table in order to clean and prevent loss of data.</p>

```python
c.execute('''
DROP TABLE IF EXISTS runner_orders_clean ;''')
c.execute('''
CREATE TABLE runner_orders_clean AS
SELECT *
FROM runner_orders

''')
conn.commit()
```

<p style='font-family:times new roman;font-size:16px;'>Fix the nulls in the table.</p>

```python
c.execute('''
UPDATE runner_orders_clean
SET pickup_time = 
    (CASE 
        WHEN pickup_time LIKE '%null%' THEN NULL
        ELSE pickup_time END)
    , distance = 
    (CASE
        WHEN distance LIKE '%null%'
            THEN NULL
        WHEN distance LIKE '%km'
            THEN TRIM(distance, 'km')
        ELSE distance END)
    , duration = 
    (CASE 
        WHEN duration LIKE '%null%'
            THEN NULL
        WHEN duration LIKE '%minutes%'
            THEN TRIM(duration, 'minutes')
        WHEN duration LIKE '%mins%'
            THEN TRIM(duration, 'mins')
        WHEN duration LIKE '%minute%'
            THEN TRIM(duration, 'minute')
        ELSE duration END)
    , cancellation =
        (CASE 
            WHEN cancellation LIKE '%null' OR cancellation = ''
                THEN NULL
            ELSE cancellation END)
    
''') 
conn.commit()
```

<p style='font-family:times new roman;font-size:16px'>Change data types for runner_orders_clean. SQLite3 does not support modifying column's data types, a new table has to be created with the appropriate data types.</p>

<h3 style='font-family:times new roman;color:navy'><center>A. Pizza Metrics</center></h3>


<p style='font-family:times new roman;font-size:16px'>1. How many pizzas were ordered?</p>

```python
pizza_orders = pd.read_sql('''
SELECT COUNT(*)as pizza_orders
FROM customer_orders_clean;
''', conn)

pizza_orders
```

<p style='font-family:times new roman;font-size:16px'>2. How many unique customer orders were made?</p>

```python
unique_orders = pd.read_sql('''
SELECT COUNT(DISTINCT order_id) AS customer_orders
FROM customer_orders_clean;

''',conn)
unique_orders
```

<p style='font-family:times new roman;font-size:16px'>3. How many successful orders were delivered by each runner?</p>

```python
successful_orders = pd.read_sql('''
SELECT runner_id
, COUNT(order_id) AS delivieries
FROM runner_orders_clean
WHERE cancellation IS NULL
GROUP BY runner_id
ORDER BY runner_id;
''',conn)
successful_orders
```

<p style='font-family:times new roman;font-size:16px'>4. How many of each type of pizza was delivered?</p>


```python
pizza_types = pd.read_sql('''
SELECT c.pizza_id
    ,COUNT(*) AS pizzas_delivered
FROM customer_orders_clean c 
JOIN runner_orders_clean r USING(order_id)
WHERE r.cancellation IS NULL
GROUP BY c.pizza_id;

''',conn)
pizza_types
```

<p style='font-family:times new roman;font-size:16px'>5. How many Vegetarian and Meatlovers were ordered by each customer?</p>


```python
types = pd.read_sql('''
SELECT c.customer_id
    , p.pizza_name
    , COUNT(c.pizza_id) AS num_pizzas_ordered
FROM customer_orders_clean c 
JOIN pizza_names p USING(pizza_id)
GROUP BY c.customer_id,p.pizza_name
ORDER BY c.customer_id, p.pizza_name
;''',conn)
types
```

<p style='font-family:times new roman;font-size:16px'>6. What was the maximum number of pizzas delivered in a single order?</p>


```python
max_order = pd.read_sql('''
WITH num_pizzas AS(

SELECT c.order_id
    , COUNT(c.pizza_id) AS num_pizzas
FROM customer_orders_clean c 
JOIN runner_orders_clean r USING(order_id)
WHERE cancellation IS NULL
GROUP BY c.order_id)

SELECT
    MAX(num_pizzas) AS max_num_pizzas FROM num_pizzas
;''',conn)
max_order
```

<p style='font-family:times new roman;font-size:16px'>7. For each customer, how many delivered pizzas had at least 1 change and how many had no changes?</p>


```python
changes = pd.read_sql('''
WITH changes AS(
SELECT c.customer_id
    , c.pizza_id
    , c.exclusions
    , c.extras
FROM customer_orders_clean c
JOIN runner_orders_clean r USING(order_id)
WHERE r.cancellation IS NULL)

SELECT customer_id
    , SUM(CASE WHEN exclusions IS NOT NULL OR extras IS NOT NULL
                THEN 1 ELSE 0 END) AS pizza_changes
    , SUM(CASE WHEN exclusions IS NULL AND extras IS NULL
                THEN 1 ELSE 0 END) AS no_pizza_changes

FROM changes
GROUP BY customer_id
ORDER BY customer_id

;''', conn)
changes
```

<p style='font-family:times new roman;font-size:16px'>8. How many pizzas were delivered that had both exclusions and extras?</p>


```python
picky_eater = pd.read_sql('''
SELECT c.order_id
    , COUNT(c.pizza_id) num_pizza
FROM customer_orders_clean c
JOIN runner_orders_clean r USING(order_id)
WHERE r.cancellation IS NULL AND
    c.exclusions IS NOT NULL AND
    c.extras IS NOT NULL
;''',conn)
picky_eater
```

<p style='font-family:times new roman;font-size:16px'>9. What was the total volume of pizzas ordered for each hour of the day?</p>


```python
volume_hour = pd.read_sql('''
SELECT strftime('%H', order_time) AS hour_24hr
    , COUNT(pizza_id) AS num_pizza
FROM customer_orders_clean
GROUP BY strftime('%H', order_time)
''',conn)
volume_hour
```

<p style='font-family:times new roman;font-size:16px'>10. What was the volume of orders for each day of the week?</p>

```python
# # SQLite3 does not support dayname(). Mannually cast the weekdays.
volume_day = pd.read_sql('''
WITH volume_day AS (
SELECT pizza_id
    , CASE CAST(strftime('%w', order_time) AS INTEGER)
            WHEN 0 THEN 'Sunday'
            WHEN 1 THEN 'Monday'
            WHEN 2 THEN 'Tuesday'
            WHEN 3 THEN 'Wednesday'
            WHEN 4 THEN 'Thursday'
            WHEN 5 THEN 'Friday'
            WHEN 6 THEN 'Saturday'
            END AS day_of_week
FROM customer_orders_clean)

SELECT COUNT(pizza_id) AS num_pizza
    , day_of_week
FROM volume_day
GROUP BY day_of_week
ORDER BY num_pizza

;''',conn)
volume_day
```

<h3 style='font-family:times new roman;color:navy'><center>B. Runner and Customer Experience</center></h3>

<p style='font-family:times new roman;font-size:16px'>1. How many runners signed up for each 1 week period? (i.e. week starts 2021-01-01)</p>

```python
runner_weekly = pd.read_sql('''
SELECT COUNT(runner_id) as num_runner
    , strftime('%W',registration_date) AS registration_week
FROM runners
GROUP BY strftime('%W', registration_date)

;''',conn)
runner_weekly
```

<p style='font-family:times new roman;font-size:16px'>2. What was the average time in minutes it took for each runner to arrive at the Pizza Runner HQ to pickup the order?</p>

```python
# # timestampdiff function not supported on sqlite3.
# # Use julianday function and multiply by 86400 to convert to seconds.
# # Then convert to minutes.
avg_time = pd.read_sql('''
SELECT  r.runner_id
    , ROUND(AVG(
    ((JULIANDAY(r.pickup_time) - JULIANDAY(c.order_time)) * 86400)/ 60)
    ,1) AS avg_time_min
FROM runner_orders_clean r
JOIN customer_orders_clean c USING(order_id)
WHERE pickup_time NOT NULL
GROUP BY runner_id
ORDER BY avg_time_min

;''',conn)
avg_time
```

<p style='font-family:times new roman;font-size:16px'>3. Is there any relationship between the number of pizzas and how long the order takes to prepare?</p>

```python
duration_pizza = pd.read_sql('''
WITH pizza_prep AS (
    SELECT  c.order_id
    , COUNT(c.order_id) as num_pizza
        , ROUND((((JULIANDAY(r.pickup_time)- JULIANDAY(c.order_time)) *86400) / 60),2)
        AS prep_duration_min
    FROM customer_orders_clean c
    JOIN runner_orders_clean r USING(order_id)
    WHERE r.cancellation is NULL 
    GROUP BY c.order_id )
    
SELECT  num_pizza
    , ROUND(AVG(prep_duration_min), 0) as avg_prep_min
FROM pizza_prep
GROUP BY num_pizza
;''',conn)
duration_pizza
```

<p style='font-family:times new roman;font-size:16px'>4. What was the average distance travelled for each customer?</p>

```python
avg_dist = pd.read_sql('''
SELECT c.customer_id
    , ROUND(AVG(r.distance_km),1) as avg_distance
FROM customer_orders_clean c
JOIN runner_orders_clean r USING(order_id)
WHERE r.cancellation IS NULL
GROUP BY c.customer_id
ORDER BY c.customer_id
;''', conn)
avg_dist
```

<p style='font-family:times new roman;font-size:16px'>5. What was the difference between the longest and shortest delivery times for all orders?</p>

```python
duration = pd.read_sql('''
SELECT MAX(duration_min) - MIN(duration_min) AS difference
FROM runner_orders_clean
WHERE cancellation IS NULL

;''',conn)
duration
```

<p style='font-family:times new roman;font-size:14px'>There is a difference of 30 minutes between the longest and shorter delivery time.</p>

<p style='font-family:times new roman;font-size:16px'>6. What was the average speed for each runner for each delivery and do you notice any trend for these values?</p>

```python
runner_speed = pd.read_sql('''

SELECT runner_id
    , order_id
    , ROUND(AVG(distance_km / (CAST(duration_min AS REAL)/ 60)),1) AS avg_speed_kmph
    , ROUND(AVG((distance_km / 1.609) / (CAST(duration_min AS REAL) / 60)),1)  AS avg_speed_mph

FROM runner_orders_clean
WHERE cancellation IS NULL
GROUP BY runner_id, order_id


;''', conn)
runner_speed
```

<p style='font-family:times new roman;font-size:16px'>7. What is the successful delivery percentage for each runner?</p>

```python
runner_pct = pd.read_sql('''
SELECT runner_id
    , SUM(CASE WHEN cancellation IS NULL THEN 1
        ELSE 0 END) / CAST(COUNT(order_id) AS REAL) * 100 AS success_pct
FROM runner_orders_clean
GROUP BY runner_id


;''', conn)
runner_pct 
```

<p style='font-family:times new roman;font-size:16px'>1. What are the standard ingredients for each pizza?</p>

```python
standard = pd.read_sql('''

SELECT n.pizza_name
    , GROUP_CONCAT(t.topping_name) as toppings
FROM pizza_recipes_mod pr
JOIN pizza_toppings t 
    ON pr.toppings = t.topping_id
JOIN pizza_names n
    ON pr.pizza_id = n.pizza_id
GROUP BY pr.pizza_id

;''', conn)
standard
```

<p style='font-family:times new roman;font-size:16px'>Split exclusions and extras field in the customer_orders_table table to make it easier to work with.</p>

<p style='font-family:times new roman;font-size:16px'>2. What was the most commonly added extra?</p>

```python
# # Functions for splitting columns in SQLITE3 are limited.
extras = pd.read_sql('''
WITH extras AS (
SELECT order_id
    , pizza_id
    , CAST(extras AS INTEGER) AS extras
FROM customer_orders_clean
WHERE LENGTH(extras) = 1
AND extras NOT NULL

UNION ALL 

SELECT order_id
    , pizza_id
    , CAST(SUBSTR(extras, 0, INSTR(extras, ',')) AS INTEGER)  AS extras
FROM customer_orders_clean
WHERE extras NOT NULL
AND LENGTH(extras)>2

UNION ALL

SELECT order_id
    , pizza_id
    , CAST(SUBSTR(extras, 3)AS INTEGER) AS extras
FROM customer_orders_clean
WHERE LENGTH(extras)>2
), 

count_extras AS(

SELECT extras,
    COUNT(extras) as count
FROM extras
GROUP BY extras
ORDER BY count DESC
LIMIT 1
) 


SELECT topping_name
FROM count_extras
JOIN pizza_toppings
    ON count_extras.extras = pizza_toppings.topping_id
;''', conn)
extras
```

<p style='font-family:times new roman;font-size:16px'>3. What was the most common exclusion?</p>

```python
# # Functions for splitting columns in SQLITE3 are limited.
exclusions = pd.read_sql('''
WITH exclusions AS (
SELECT order_id
    , pizza_id
    , CAST(exclusions AS INTEGER) AS exclusions
FROM customer_orders_clean
WHERE LENGTH(exclusions) = 1
AND exclusions NOT NULL

UNION ALL 

SELECT order_id
    , pizza_id
    , CAST(SUBSTR(exclusions, 0, INSTR(exclusions, ',')) AS INTEGER)  AS exclusions
FROM customer_orders_clean
WHERE exclusions NOT NULL
AND LENGTH(exclusions)>2

UNION ALL

SELECT order_id
    , pizza_id
    , CAST(SUBSTR(exclusions, 3)AS INTEGER) AS exclusions
FROM customer_orders_clean
WHERE LENGTH(exclusions)>2
), 

count_exclusions AS(

SELECT exclusions,
    COUNT(exclusions) as count
FROM exclusions
GROUP BY exclusions
ORDER BY count DESC
LIMIT 1
) 


SELECT topping_name
FROM count_exclusions
JOIN pizza_toppings
    ON count_exclusions.exclusions = pizza_toppings.topping_id
;''', conn)
exclusions
```
